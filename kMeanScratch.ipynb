```
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import numpy as np
from math import sqrt
import random
#from sklearn import preprocessing
```

```
def euclidean(p1,p2):
    sum=0
    for i in range(len(p1)):
        #print(p1[i],p2[i])
        sum+=abs(float(p1[i])-float(p2[i]))*abs(float(p1[i])-float(p2[i]))
        #print(sum)
    return sqrt(sum)
```

```
## to calculate mean
def mean(group):
    ans=[]
    for j in range(len(group[0])):
        temp=0
        for i in range(len(group)):
            temp+=group[i][j]
        ans.append(temp//len(group[0]))

    return ans
```

```
class K_Means:
  def __init__(self,k=2,tol =0.001,max_iter=300):
    self.k=k
    self.tol=tol
    self.max_iter=max_iter
  def fit(self,data):
    self.centroids={}
    group= random.sample(list(data),self.k)
    for i in range(self.k):
      self.centroids[i]=group[i]
    for i in range(self.max_iter):
      self.classifications={}
      #assigning empty list to dictionary with key as centroid id or cluster id
      for i in range(self.k):
        self.classifications[i]=[]
      for featureset in data:
        distances =[euclidean(featureset,self.centroids[centroid]) for centroid in self.centroids]
        clusterID = distances.index(min(distances))
        self.classifications[clusterID].append(featureset)
      prev_centroids = dict(self.centroids)
      for clusterID in self.classifications:
        self.centroids[clusterID]= mean(self.classifications[clusterID])
      
      optimized =True
      for c in self.centroids:
        original_centroid = prev_centroids[c]
        current_centroid = self.centroids[c]
        #try to implement sum() here ,try not to use numpy
        if np.sum((current_centroid-original_centroid)/original_centroid*100.0)>self.tol:
          optimized = False
        
      if optimized:
        break




  def predict(self,data):
    distances =[euclidean(featureset,self.centroids[centroid]) for centroid in self.centroids]
    clusterID = distances.index(min(distances))
    return clusterID
```

```
X= np.array([[2,22],
            [2,1],
            [4,5],
            [2,1],
            [1,3],
            [8,9],
            [0,3],
            [5,4],
            [6,4],
            [1,3]])
```

```
clf = K_Means()
clf.fit(X)
```

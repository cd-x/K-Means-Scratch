# **Preprocessing Begins Here**

**Dropping attributes which does not affect the result**

```
df.drop(['Name','Ticket','Cabin'],1,inplace=True)
```

```
# embark={'C':0,'Q':1,'S':2}
# for each in df['Embarked']:
#   each.replace(embark['each'])
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
```

```
#df.replace({'Embarked': {'S': 0, 'C': 1,'Q': 2}})
```

```
df=df.fillna(method ='pad')
df=df.replace({'Embarked': {'S': 0, 'C': 1,'Q': 2}})
```

```
# df.drop(['Survived'],1,inplace=True)
```

Importing packages

```
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import numpy as np
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

Scaling Data for better result (like normalization)

```
#X = np.array(df.drop(['Survived'],1).astype(float))
X= np.array(df.astype(float))
X = preprocessing.scale(X)
#y= np.array(df['Survived'])
```

Fitting training data

```
clf = KMeans(n_clusters=2)
clf.fit(X)
```

**Preprocessing test dataset below**

```
dfTest= pd.read_csv('/gdrive/My Drive/Colab Notebooks/titanic-train.csv')
dfTest.drop(['Name','Ticket','Cabin'],1,inplace=True)
dfTest['Sex']=le.fit_transform(dfTest['Sex'])
dfTest=dfTest.fillna(method ='pad')
dfTest=dfTest.replace({'Embarked': {'S': 0, 'C': 1,'Q': 2}})
```

Getting the test set and y(survival value 1 or 0)

```
X_test = np.array(dfTest.drop(['Survived'],1).astype(float))
#X= np.array(dfTest.astype(float))
X_test = preprocessing.scale(X_test)
y= np.array(dfTest['Survived'])
```

Checking accuracy

```
correct = 0
for i in range(len(X_test)):
  predict_this = np.array(X_test[i].astype(float))
  predict_this=predict_this.reshape(-1,len(predict_this))
  prediction = clf.predict(predict_this)
  if(prediction[0]!=y[i]):
    correct+=1
```

**The accurarcy is :**

```
print(correct/len(X_test))
```
